# Calcular la varianza usando Map Reduce

```python
import urllib

f = urllib.urlretrieve ("http://kdd.ics.uci.edu/databases/kddcup99/kddcup.data_10_percent.gz", "kddcup.data_10_percent.gz")
```

```python
from pyspark import SparkContext
sc=SparkContext('local','Simple App')
data_file = "./kddcup.data_10_percent.gz"
raw_data = sc.textFile(data_file)
```

```python
raw_data.count()
```

```python
#Varianza usando map reduce
```

```python
raw_data_sample = raw_data.sample(False, 0.0021, 1234).map(lambda x: int(x.split(",")[0])).map(lambda x:[x**2,x,1])
```

```python
raw_data_sample.collect()
```

```python
normal_sum_count = raw_data_sample.reduce(lambda x,y:[x[0]+y[0],x[1]+y[1],x[2]+y[2]])
normal_sum_count
```

```python
#Varianza
normal_sum_count[0]/normal_sum_count[2]-(normal_sum_count[1]/normal_sum_count[2])**2
```
